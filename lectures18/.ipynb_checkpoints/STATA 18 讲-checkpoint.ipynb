## 命令

### 命令格式

```stata
// [by varlist:] command [varlist] [=exp] [if exp] [in range] [weight] [,options]
```

1. [by varlist] ：分类操作

2. command [varlist] ：命令

3. [=exp] ：赋值及运算表达式

4. [if exp] ：条件表达式

5. [in range] ：范围筛选

6. [weight] ：加权
 
7. [,options] ：其他可选项

### 常用命令

```stata
* 帮助查找
help
search


* 打开和保存数据
clear
use
save

* 导入数据
input
edit
infile
insheet

* 重整数据
append
merge
xpose
reshape
generate
egen//?
rename
drop
keep
sort
encode
decode
order
by

* 报告数据
describe
codebook
list
count
inspect
table
tabulate

* 显示和保存结果
display
log
```

### 案例3:横向连接数据文件

```stata
chdir /Users/mac/MyPrograme/Stata资料/data
use economy,clear
sort id
save economy,replace
use math,clear
sort id
merge id using economy
tab_merge  显示对接情况
drop_merge
sort id
save m_e, replace
use students,clear
sort id
merge id using m_e
tab_merge  显示对接情况
drop_merge
save mymerge
```
